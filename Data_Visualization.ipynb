Create visualizations to represent the distribution of ratings using different charts (histogram, bar plot, etc.).

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import pandas as pd
from google.colab import files
import os

#Upload the csv file
uploaded = files.upload()

#Get the correct filename
file_name = list(uploaded.keys())[0]  # Automatically get the uploaded file's name

# Read the CSV file
df = pd.read_csv(file_name)

# check if the dataset is loaded correctly
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set the plot style
sns.set_style("whitegrid")

#Create subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18,5))


# Histogram (Distribution of Ratings)
sns.histplot(df["Aggregate rating"], bins=10, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Distribution of Ratings")
axes[0].set_xlabel("Aggregate rating")
axes[0].set_ylabel("Frequency")

# Bar Plot (Count of Different Rating Texts)
sns.countplot(data=df, x="Rating text", order=df["Rating text"].value_counts().index, ax=axes[1], palette="viridis")
axes[1].set_title("Count of Different Rating Texts")
axes[1].set_xlabel("Rating Text")
axes[1].set_ylabel("Count")

# Box Plot (Spread of Ratings)
sns.boxplot(data=df, x="Rating text", ax=axes[2], color="red")
axes[2].set_title("Spread of Ratings")
axes[2].set_xlabel("Rating Text")

# Show all plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(file_name)

# Display first few rows
print(df.head())
```

# **Compare Average Ratings using a Bar Plot**

**Bar Plot for Average Ratings by Cuisine**

```python
# Compute average ratings for each cuisine
avg_ratings_cuisine = df.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(x=avg_ratings_cuisine.index, y=avg_ratings_cuisine.values, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Cuisines")
plt.ylabel("Average Rating")
plt.title("Average Ratings by Cuisine")
plt.show()
```

**Bar Plot for Average Ratings by City**

```python
# Compute average ratings for each city
avg_ratings_city = df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(x=avg_ratings_city.index, y=avg_ratings_city.values, palette="coolwarm")
plt.xticks(rotation=45, ha="right")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.title("Average Ratings by City")
plt.show()
```

**Box Plot for Cuisine Ratings**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Cuisines", y="Aggregate rating", data=df, palette="Set3")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Cuisines")
plt.ylabel("Aggregate rating")
plt.title("Ratings Distribution by Cuisines")
plt.show()
```

**Box Plot for City Ratings**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="City",y="Aggregate rating", data=df, palette="coolwarm")
plt.xticks(rotation=45, ha="right")
plt.xlabel("City")
plt.ylabel("Aggregate rating")
plt.title("Ratings Distributions by City")
plt.show()
```

**Visualize the relationship between various features and the target variable to gain insights**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#upload the csv file
uploaded = files.upload()

#Get the correct filename
file_name = list(uploaded.keys())[0] # Automatically get the uploaded file's name

# Read the CSV file
df = pd.read_csv(file_name)

#check if the dataset is loaded correctly
print(df.head())
```

**Scatter Plot - Relationship Betwwen Cost and Rating**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df["Average Cost for two"], y=df["Aggregate rating"], alpha=0.5, color="green")
plt.xlabel("Average Cost for Two")
plt.ylabel("Aggregate Rating")
plt.title("Average Cost for Two vs Aggregate Rating")
plt.show()
```

**Box Plot-Price Range vs Rating**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df["Price range"], y=df["Aggregate rating"], palette="coolwarm")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs Aggregate Rating")
plt.show()
```

**Bar Plot - Impact of Table Booking on Rating**

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df["Has Table booking"], y=df["Aggregate rating"], palette="viridis")
plt.xlabel("Has Table Booking")
plt.ylabel("Aggregate Rating")
plt.title("impact of Table Booking on Rating")
plt.show()
```

**Bar Plot - Impact of Online Delivery on Rating**

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df["Has Online delivery"], y=df["Aggregate rating"], palette="coolwarm")
plt.xlabel("Has Online Delivery")
plt.ylabel("Aggregate Rating")
plt.title("Impact of online Delivery on Aggregate Rating")
plt.show()
```

**City-Wise Average Ratings**

```python
plt.figure(figsize=(12,6))
city_ratings = df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False)
sns.barplot(x=city_ratings.index, y=city_ratings.values, palette="Set3")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.title("city-wise Average Ratings")
plt.show()
```

```python
plt.figure(figsize=(12,6))
city_ratings = df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=city_ratings.index, y=city_ratings.values, palette="Set3")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.title("Top 10 cities with Highest Average Ratings")
plt.xticks(rotation=45, ha="right")
plt.show()
```

**Cuisine-Wise Average Ratings**

```python
plt.figure(figsize=(12,6))
cuisine_ratings = df.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending=False)
sns.barplot(x=cuisine_ratings.index, y=cuisine_ratings.index, palette="plasma")
plt.xlabel("Average Rating")
plt.ylabel("Cuisines")
plt.title("Cuisine-wise Average Ratings")
plt.show()
```

```python
plt.figure(figsize=(12,6))
cuisine_ratings = df.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=cuisine_ratings.index, y=cuisine_ratings.index, palette="plasma")
plt.xlabel("Average Rating")
plt.ylabel("Cuisines")
plt.title("Cuisine-wise Average Ratings")
plt.show()
```
