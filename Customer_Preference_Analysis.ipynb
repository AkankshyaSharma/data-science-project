## **Analyze the relationship between the type of cuisine and the restaurant's rating.**

```python
import pandas as pd
from google.colab import files
uploades = files.upload()

# Upload the csv file
uploaded = files.upload()

# Get the correct filename
filename = list(uploaded.keys())[0]      # Automatically get the uploaded file's name

# Read the CSV file
df = pd.read_csv(filename)

# check if the dataset is loaded correctly
print(df.head())
```

**Select relevant columns**

```python
# Selecting relevant columns and dropping rows with missing values
df_cuisine = df[['Cuisines' ,'Aggregate rating' ]].dropna()

# Displaying first few rows
df_cuisine.head()
```

**Handle multiple cuisines**

```python
# Splitting multiple cuisines into separate rows
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.split(',')
df_exploded = df_cuisine.explode('Cuisines')

# Displaying first few rows after transformation
df_exploded.head()
```

**Calculate average rating for each cuisine**

```python
# Convert ratings to numeric (avoid errors in sorting)
df_exploded['Aggregate rating'] = pd.to_numeric(df_exploded['Aggregate rating'], errors='coerce')

# Group by cuisine type and calculate the mean rating
cuisine_ratings = df_exploded.groupby('Cuisines', as_index=False)['Aggregate rating'].mean()

# Sort cuisines by highest rating
cuisine_ratings_sorted = cuisine_ratings.sort_values(by='Aggregate rating', ascending=False)

# Display the top 10 highest-rated cuisines
print(cuisine_ratings_sorted.head(10))
```

## **Identify the most popular cuisines among customers based on the number of votes.**

```python
import pandas as pd
from google.colab import files
uploades = files.upload()

# Upload the csv file
uploaded = files.upload()

# Get the correct filename
filename = list(uploaded.keys())[0]      # Automatically get the uploaded file's name

# Read the CSV file
df = pd.read_csv(filename)

# check if the dataset is loaded correctly
print(df.head())
```

**Select relevant columns**

```python
# Select only necessary columns and drop missing Values
df_cuisine = df[['Cuisines','Votes']].dropna()

# Display first few rows
df_cuisine.head()
```

**Handle multiple cuisines**

```python
# Splitting multiple cuisines into separate rows
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.split(',')
df_exploded = df_cuisine.explode('Cuisines')

# Displaying first few rows after transformation
df_exploded.head()
```

**Calculate total votes per cuisine**

```python
# calculate total votes per cuisine and sort
cuisine_popularity = df.groupby('Cuisines')['Votes'].sum().reset_index()
cuisine_popularity = cuisine_popularity.sort_values(by='Votes',ascending=False)

# Show top 10 most popular cuisines
print(cuisine_popularity.head(10))
```
